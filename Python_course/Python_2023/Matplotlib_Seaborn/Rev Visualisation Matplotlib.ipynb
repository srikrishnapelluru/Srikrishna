## Visualising data

When working with data, whether to draw insights or to build predictive models, the human element of visualising and understanding the data is important. 

Before being able to build Machine Learning models, in most real-life cases data requires a lot of manipulation and structuring. Missing values need to be filled, duplicates need to be removed, redundant columns(that provide no additional insight from other columns already present), new features (columns) which are derived from already existing ones need to be created and a lot of other processes are required. 

Integral to all this - while data manipulation and featuring is going on - to decide how to handle this data - and after - to see the results and determine further steps - is Data Visualisation.

Two of the most popular libraries in Python for data visualisation are matplotlib and Seaborn. Though a library of its own, the underlying foundation of Seaborn is matplotlib. Seaborn adds functionality and ease of use to matplotlib by enclosing many of the matplotlib functions in wrappers. Also, Seaborn integrates better with the most widely used data handling library in Python - Pandas. 

Before we move on to working with matplotlib library, we need to cover the types of data we will use and the appropriate visualisation techniques for them.

##### Our Data will be divided into two main types:

1. Quantitative variables. Also called numerical variables. Examples of these are number of students in a class room, heights, weights, temperature, population count etc.

2. Qualitative variables. Also called categorical variables. Examples of these are gender, citizenship, education level etc.

##### Broadly Quantitative variables are divided into two subgroups - 

Discrete - where the count is finite e.g. runs scored in a cricket match, vote count, number of students in class etc. 

Continuous - where the values can be infinite - only limited by our capability and willingness to measure/calculate. Pi value - is continuous. Heights, weights, temperature, percentage of population. A continuous value may end up evaluating to a finite value, but intrinsically, it can take on an innumerable number of values and cannot be subjected to a count. 

##### Qualitative or categorical variables are also divided further into two subgroups - 

Ordinal - where the variables have some intrinsic order. E.g. Feedback ratings - stars 1 to 5. We know that 5 star is better than 1 star. Or socio-economic status - low income, middle class, high income or education level - High school, Graduate, Post Graduate etc. 

Nominal - where the variables have no order at all - examples - race, gender and nationality. 

Nominal variables - if they have only two possible distinctions are labeled Binary Nominal variables - e.g. Yes/No, Male/Female, Black/White etc. 

Now that we know the types of variables, let us look at a few examples of different types of charts used for visualisation.

### How to Select a Meaningful Visualization

The three steps in the data visualization process are preparing, visualizing, and styling data. 

When planning out a visualization, youll usually have an idea of what questions you will want to explore. However, you may initially wonder exactly which chart to use.

You should consider two things:
-	The focusing question you want to answer with your chart.
-	Type of data that you want to visualize.

Depending on the focusing questions you're trying to answer, the type of chart you select should be different and intentional in its difference. 

#### Chart categories

##### Composition charts

Focusing Question: What are the parts of some whole? What is the data made of? 

Datasets that work well: Data pertaining to probabilities, proportions, and percentages can be visualized as with the graphs in this composition category.

Charts in this category illustrate the different data components and their percentages as part of a whole.
Type of charts  Pie Charts, Stacked Bar, Line Graphs with shaded Area Under Curve

##### Distribution Charts

Focusing question: How are the datapoints behaving with respect to statistical calculations such as mean, median, mode, standard deviation etc?

Datasets that work well: Data in large quantities and / or with an array of attributes works well for these types of charts. Visualizations in this category will allow you to see patterns, re-occurrences, and a clustering of data points.

Note: In statistics, a commonly seen distribution is a bell curve, also known as a normal distribution. A bell curve is a bell-shaped distribution where most of the values in the dataset crowd around the average (also known as the mean) therefore causing the curve to form. If you want to see how values in the data are distributed across variables, the best way to do that would be with the visualizations in this category.

Type of charts  Histograms and Bar Charts, Box Plots, Swarm Plots, Violin Plots, KDE (Kernel Density Estimator  for estimated probability density)

##### Relationship Charts

Focusing Question: How do variables relate to each other? 

Datasets that work well: Data with two or more variables can be displayed these charts. These charts typically illustrate a correlation between two or more variables. You can communicate this relationship by mapping multiple variables in the same chart. Correlation measures the strength of a relationship between variables.

Type of charts  Scatter Plots

##### Comparison Charts

Focusing Question: How do variables compare to each other? 

Datasets that work well: Data must have multiple variables, and the visualizations in this category allow readers to compare those items against the others. For example, a line graph that has multiple lines, each belonging to a different variable. Multi-coloured bar charts are also a great way to compare items in data.

Type of charts  Line graphs, Bar Graphs, KDE(Kernel Density Estimator) graphs.

#### Pie-Charts are used to display elements of a dataset as proportions of a whole. 

```python
pie_data = [10,20,30,40,50]
labels = ['Ten','Twenty', 'Thirty', 'Forty', 'Fifty']

plt.figure(figsize = (12,12))
plt.pie(pie_data, labels = labels)
plt.show()
```

##### plt.pie() with additional parameters.

```python
plt.figure(figsize = (16,10))
plt.pie(pie_data, labels = labels, colors = ['red', 'blue', 'yellow', 'green', 'orange'],autopct = "%.5f%%", \
pctdistance = 0.5, shadow = True, labeldistance = 0.75, explode = [0.1,0.0,0.0,0.2,0.0], \
rotatelabels = True, textprops = {'fontsize' : 20, 'color' : 'black', 'fontweight':100}, startangle = 90,\
wedgeprops = {'linewidth' : 2, 'edgecolor' : 'k'}, counterclock = False)
plt.legend()
plt.show()

#explode = [0.05,0.07,0.09,0.11,0.12]
```

explode : Optional parameter which takes an array or list-like input specifying how much to 'explode' (bring away from center of circle) each wedge of the pie. 

colors : Optional parameter which takes an array-like or list-like input specifying the colors for each wedge.

autopct : Show the percentage of each wedge (since it is a composition chart showing percentage of whole). Takes an input as a string format i.e. '%.2f' for formatting strings. Or '%.2f%%' to include the percentage sign while displaying the percentage. 

pctdistance : Default 0.6. How far from center (as a ratio of the circle radius) we wish our percentages to display.

labeldistance : Default 1.1. How far from center (as a ratio of the circle radius) we wish the labels to be displayed. A ratio below one brings the labels into the wedges and higher than one takes them out of the wedges. 

shadow : Default false. If set to True shows a shadow for each wedge.
