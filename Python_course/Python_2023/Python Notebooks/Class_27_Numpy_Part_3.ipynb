# Vector stacking --> Concatenation

```python
a = np.array([['bangalore' , 'Tamil Nadu'] , ['Delhi' , 'Mumbai']])

a
```

```python
b = np.array([['kerala' , 'UP'] , ['AP' , 'Pune']])

b
```

```python
np.concatenate((a,b)) # axis=0 --> Row wise 
```

```python
np.concatenate((a,b) , axis = 1) # axis=1 --> column wise 
```

```python
np.hstack((a,b)) # Column wise
```

```python
np.vstack((a,b)) # Row wise 
```
