# Slicing

- The op if slicing will always give you a new dataframe

```python
new_df = df.head(10)

e_list = []

for i in new_df.index:
    e_list.append(f'row_{i}')
    
new_df.index = e_list

new_df
```

```python
# 1. Normal slicing
- Slicing is possible only on rows not on column

* zero indexing
- start is inclusive
- stop is exclusive

* custom indexing 
- start is inclusive
- stop is inclusive


# 2. loc method

- Slicing on column is also possible

syntax
dataframe.loc[rows , column]

* custom indexing 
- start is inclusive
- stop is inclusive


# 3. iloc method

- Slicing on column is also possible

syntax
dataframe.iloc[rows , column]

* zero indexing
- start is inclusive
- stop is exclusive
```
