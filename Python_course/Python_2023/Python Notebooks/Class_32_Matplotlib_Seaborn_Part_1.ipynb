```python
days=[1,2,3,4,5,6,7] # weekdays
max_t=[50,51,52,48,47,49,46] # max temp recorder at that day
min_t=[43,42,40,44,45,35,37] # min temp
avg_t=[45,48,48,46,40,42,41] # avg temp
```

```python
plt.plot(days , max_t)
plt.plot(days , min_t)
plt.plot(days , avg_t)

plt.title('Weather info')
plt.xlabel('DAY')
plt.ylabel('Temp')
```

```python
plt.plot(days , max_t  ,label = 'MAX')
plt.plot(days , min_t , label = 'MIN')
plt.plot(days , avg_t , label = 'AVG')

plt.title('Weather info')
plt.xlabel('DAY')
plt.ylabel('Temp')
plt.legend()
```

```python
plt.plot(days , max_t  ,label = 'MAX')
plt.plot(days , min_t , label = 'MIN')
plt.plot(days , avg_t , label = 'AVG')

plt.title('Weather info')
plt.xlabel('DAY')
plt.ylabel('Temp')
plt.legend(loc ='lower left')
```

```python
plt.plot(days , max_t  ,label = 'MAX')
plt.plot(days , min_t , label = 'MIN')
plt.plot(days , avg_t , label = 'AVG')

plt.title('Weather info')
plt.xlabel('DAY')
plt.ylabel('Temp')
plt.legend(loc =6)
```

```python
==================   =============
        Location String      Location Code
        ==================   =============
        'best' (Axes only)   0
        'upper right'        1
        'upper left'         2
        'lower left'         3
        'lower right'        4
        'right'              5
        'center left'        6
        'center right'       7
        'lower center'       8
        'upper center'       9
        'center'             10
        ==================   =============
```

```python
plt.plot(days , max_t ,'o--'  ,label = 'MAX')
plt.plot(days , min_t,'<:', label = 'MIN')
plt.plot(days , avg_t,'>-', label = 'AVG')

plt.title('Weather info')
plt.xlabel('DAY')
plt.ylabel('Temp')
plt.grid()
```

```python
blood_sugar = [113 , 85,90,150 , 149 , 88 , 93 , 115, 135 , 80 , 77 , 82 , 129]
```

```python
80 - 100 ----> Normal
100 - 125 ---> pre diabetic
125 and above --> diabetic
```

```python
plt.hist(blood_sugar)
```

```python
blood_sugar =[113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]


plt.hist(blood_sugar , bins= [77 , 100 , 125 , 150])
```

```python
blood_sugar =[113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]


plt.hist(blood_sugar , bins= [77 , 100 , 125 , 150])
# plt.text(85 ,5 , 'N')
# plt.text(110 ,1 , 'P')
# plt.text(140 ,1 , 'D')
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]


plt.hist([blood_sugar_men , blood_sugar_women] , bins= [67 , 100 , 125 , 150] , label= ['Men' , 'Women'])
plt.legend()
```

# Pie

```python
exp_values = [1400 , 600 , 300, 410 , 250]
Values_spent_on = ['Rent' , 'Food' , 'mobile' , 'Car' , 'Others']
```

```python
plt.pie(exp_values)
plt.show()
```

```python
plt.pie(exp_values , labels= Values_spent_on)
plt.show()
```

```python
plt.pie(exp_values , labels= Values_spent_on , radius= 2)
plt.show()
```

```python
plt.pie(exp_values , labels= Values_spent_on ,autopct= '%0.2f%%')


plt.show()
```

```python
exp_values = [1400 , 600 , 300, 410 , 250]
Values_spent_on = ['Rent' , 'Food' , 'mobile' , 'Car' , 'Others']

plt.pie(exp_values , labels= Values_spent_on ,autopct= '%0.2f%%' , explode = [0,0,0.5,0,0])


plt.show()
```

```python
exp_values = [1400 , 600 , 300, 410 , 250]
Values_spent_on = ['Rent' , 'Food' , 'mobile' , 'Car' , 'Others']

plt.pie(exp_values , labels= Values_spent_on ,autopct= '%0.2f%%' , explode = [0,0,0.5,0,0.25])


plt.show()
```
